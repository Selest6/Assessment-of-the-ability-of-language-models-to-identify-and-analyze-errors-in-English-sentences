```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

#        
def process_files(file_paths):
    results = []
    for file_path in file_paths:
        df = pd.read_csv(file_path)

        #       
        file_name = os.path.splitext(os.path.basename(file_path))[0]

        #     "total (out of...)"  "structure (out of...)"
        total_col = [col for col in df.columns if col.startswith('total (out of')][0]
        structure_col = [col for col in df.columns if col.startswith('structure (out of')][0]
        correct_exp2_col = [col for col in df.columns if col == 'correct_exp2'][0]

        #     
        df[total_col] = pd.to_numeric(df[total_col], errors='coerce')
        df[structure_col] = pd.to_numeric(df[structure_col], errors='coerce')
        df[correct_exp2_col] = pd.to_numeric(df[correct_exp2_col], errors='coerce')

        total_sum = df[total_col].sum() / 400 # 8*50
        structure_sum = df[structure_col].sum() / 200 # 4*50
        content_sum = df[correct_exp2_col].sum() / 200

        results.append((file_name, total_sum, content_sum, structure_sum))

    return results

#    (   )
file_paths = ['label explanation correction.csv', 'label category explanation correction.csv']

#  
results = process_files(file_paths)

#  DataFrame   
heatmap_data = pd.DataFrame(results, columns=['File', 'Explanation and correction', 'Content', 'Structure'])
heatmap_data.set_index('File', inplace=True)

#   
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar_kws={'label': 'Value'})
plt.title('Heatmap of Explanation and Correction (including structure and content) Values')
plt.xlabel('Metrics')
plt.ylabel('Files')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

#        
def process_files(file_paths):
    results = []
    for file_path in file_paths:
        df = pd.read_csv(file_path)

        #       
        file_name = os.path.splitext(os.path.basename(file_path))[0]

        #    "
        total_col = [col for col in df.columns if col.startswith('is category')][0]

        #     
        df[total_col] = pd.to_numeric(df[total_col], errors='coerce')

        total_sum = df[total_col].sum() / 50 # 1*50

        results.append((file_name, total_sum))

    return results

#    (   )
file_paths = ['label category.csv', 'label category explanation.csv', 'label category correction.csv', 'label category explanation correction.csv']

#  
results = process_files(file_paths)

#  DataFrame   
heatmap_data = pd.DataFrame(results, columns=['File', 'Category'])
heatmap_data.set_index('File', inplace=True)

#   
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar_kws={'label': 'Value'})
plt.title('Heatmap of Category Values')
plt.xlabel('Metrics')
plt.ylabel('Files')
plt.show()
```
